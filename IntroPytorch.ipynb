# **AUTOGRAD**

```python
x= torch.tensor(1.,requires_grad=True)
y= torch.tensor(2.,requires_grad=True)
p= x+y

z = torch.tensor(3.,requires_grad=True)

g = p * z
```

```python
g.backward()
```

```python
z.grad
```

```python
x.grad
```
