```python
import torch
#creamos una clase que hereda de troch.nn.Module  como se hace a mano

class Model(torch.nn.Module):

  #constructor
  def __init__(self,D_in,H,D_out):
  
  #llamamaos al constructor de la clase madre
    super(Model,self).__init__()

    #definimos las capas

    self.fc1=torch.nn.Linear(D_in,H)
    self.relu = torch.nn.ReLU()
    self.fc2 = torch.nn.Linear(H,D_out)

  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x
  

```

```python
model = Model(784,100,10)
outputs = model(torch.randn(64,784))
outputs.shape
```
