Definiendo la primera clase

```python
class miclase:

  #constructor
  def __init__(self,a): #para poder inicializar el valor de las variables como queramos y no sea fija
    self.a=a

  #sobrecarga suma
  def __add__(self,other): #(un objeto de la clase, otro objeto de la misma clase)
    return self.a + other.a


  def func(self):  # self indica que es de su misma clase
    print(self.a)
```

```python
#Creando un objeto de esta clase creada
x= miclase("holis") # x pasa a ser un objeto de nuestra clase
x.func
```

```python
x.a
```

```python
#siqueremos llamr a la funcion
x.func()
```

```python
#modificamos un atributo

x.a= "modificando el atributo"
x.a
```

generadores

```python
class miiterador:
  #constructor
  def __init__(self,n):
    self.items = [i for i in range(n)]

  # longitud de la secuencia
  def __len__(self):
    return len(self.items)

  # get item
  def __getitem__(self,ix): 
    return self.items[ix]
     
```

```python
x= miiterador(5)

for t in x:
  print(t)

""
""

len(x)
```

```python
class MiAlgoritmoBase:
    a = 1    
    def __call__(self):
        return 2*self.a
```

```python
class MiAlgoritmo(MiAlgoritmoBase):
    
    def __init__(self, b):
        self.b = b
    
    def __call__(self):
        # el atributo `a` viene de la clase madre
        return 2*self.a + self.b
```

```python
x = MiAlgoritmoBase()

x()
```

```python
x = MiAlgoritmo(2)

x()
```
