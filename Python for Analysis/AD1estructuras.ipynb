```python
#armando dicc  con 2 listas de valores y claves
d={}
keys=['a','b','c']
values=[1,2,3]

for k,v in zip(keys,values):
  d[k]=v

print(d)
```

```python
##############-------list comprehension-----------########### muy uilites para crear estructuras de datos a partir de otras

l = [1,2,3,4,5]
#obteniendo un nueva lista con elementos multiplicados por 2 DED MANERA CLASICA

l2= []
for i in l:
  l2.append(i*2)

l2  
```

```python
#armando un dict interesante
l1= ["a","b","c"]
l2 =[1,2,3]

d= {k:v for k, v in zip(l1,l2) if v <= 2}
d
```
