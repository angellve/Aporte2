tareas en secuencial

```python
import time
p=[1,2,3,4]  

def func(p):
  print(f"soy el proceso{p}")
  time.sleep(1) #se queda quieto durante 1 segundo
  return 2*p



% time results=[func(_p) for _p in p] #cuanto tarde en ejecutarse esa line
results
```

```python
#mejorando la secuencialidad a paralelo
import concurrent.futures

with concurrent.futures.ThreadPoolExecutor() as executor:
  %time results = list(executor.map(func,p))

results  
```
