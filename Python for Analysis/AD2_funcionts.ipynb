```python
#funcion de varios argumentos
def func(a,b,c="hola"): #por defecto c va a hacer hola,siempre dejamos los opcionales al final de la funcion
  print(a,b,c)

```

```python
func("hola","que","tal")
```

```python
#probando con solo dos argumentos
func("hola","que")
```

```python
#tambien podemos usarlo con diferentes argumentos por su nombre, sin orden alguno
func(b=1, a=2)
```

```python
#la funcion suma y resta
def funt(k,h):
  return k + h, k-h #devolvera una tupla
```

```python
#guardando la funcion en una variable
x1,x2= funt(5,2)

x2
```

```python
#------generadores--------# Nos permite iterar sobre secuencias de manera cumtom

def funz():
  for i in range(10):
    if i %2:
      yield i
```

```python
#instancionado generador
gen= funz()
for i in gen:
  print(i)
```

```python
#realizando generadores con listcomprehasion

gen= (i for i in range(10) if i % 2)

for i in gen:
  print(i)
```
